```python
import sys
class Node:
    def __init__(self, val):
        self.value = val
        self.left = None
        self.right = None
class Tree_:
    def __init__(self, root = None):
        self.root = root
    def get_(self, key):
        return self._get_(self.root, key)
    def _get_(self, node, key):
        if node is None:
            return None
        elif node.value == key:
            return key
        else:
            if key < node.value:
                return self._get_(node.left, key)
            else:
                return self._get_(node.right, key)
    def get_max(self):
        return self._get_max(self.root)
    def _get_max(self, node):
        if node.right is None:
            return node.value
        else:
            return self._get_max(node.right)
    def add(self, key):
        self.root = self._add(self.root, key)
    def _add(self, node, key):
        if node is None:
            return Node(key)
        elif node.value == key:
            pass
        else:
            if key < node.value:
                node.left = self._add(node.left, key)
            else:
                node.right = self._add(node.right, key)
        return node  
    def remove(self, key):
        self.root = self._remove(self.root, key)
    def _remove(self, node, key):
        if node is None:
            return None
        elif key < node.value:
            node.left = self._remove(node.left, key)
        elif key > node.value:
            node.right = self._remove(node.right, key)
        else:
            if node.left is None:
                node = node.right
            elif node.right is None:
                node = node.left
            else:
                node.value = self._get_max(node.left)
                node.left = self._remove(node.left, node.value)
        return node
    def tree_size(self):
        return self._size(self.root)
    def _size(self, node):
        if node is None:
            return 0
        else:
            return 1 + self._size(node.left) + self._size(node.right)
    def max_depth(self):
        return self._max_depth(self.root)
    def _max_depth(self, node):
        if node is None:
            return 0
        else:
            left = self._max_depth(node.left)
            right = self._max_depth(node.right)
            return max(left, right) + 1
    def min_depth(self):
        return self._min_depth(self.root)
    def _min_depth(self, node):
        if node is None:
            return 0
        else:
            left = self._min_depth(node.left)
            right = self._min_depth(node.right)
            return min(left, right) + 1
    def isBalance(self):
        return (self.max_depth() - self.min_depth()) <= 1
    def get_floor(self, key):  # 找比key小的第一个值
        return self._get_floor(self.root, key)
    def _get_floor(self, node, key):
        if node is None:
            return None  # 如果这个节点什么都没有就直接结束
        if node.value == key:
            return node
        if key < node.value:  # 如果大于key就去左边找，而且这个点根本没法用，所以不用存
            return self._get_floor(node.left, key)
        t = self._get_floor(node.right, key)  # 如果小于key就去右边再看一下有没有更贴近一点的，但是把这点存下来
        if t:  # 万一去右边找到了更大一点但是又小于key的那就接受
            return t
        return node  # 没找到就返回当前这个点
        

class Solution:
    def pre_order(self, root):
        ans = []
        self._pre_order(root, ans)
        return ans
    def _pre_order(self, node, ans=[]):
        if node:
            ans.append(node.value)
        if node.left:
            self._pre_order(node.left, ans)
        if node.right:
            self._pre_order(node.right, ans)
#         return ans
    def in_order(self, root):
        ans = []
        self._in_order(root, ans)
        return ans
    def _in_order(self, node, ans=[]):
        if node.left:
            self._in_order(node.left, ans)
        if node:
            ans.append(node.value)
        if node.right:
            self._in_order(node.right, ans)
#         return ans
    def post_order(self, root):
        ans = []
        self._post_order(root, ans)
        return ans
    def _post_order(self, node, ans=[]):
        if node.left:
            self._post_order(node.left, ans)
        if node.right:
            self._post_order(node.right, ans)
        if node:
            ans.append(node.value)
#         return ans
    def isBST(self, root): # 检查是否是一棵二叉搜索树
        return self._isBST(root, -sys.maxsize, sys.maxsize)
    def _isBST(self, node, minval, maxval):  # 思想就是检查每个点的大小范围是否满足要求，大于或等于左右的所有点
        if node is None:
            return True
        if node.value < minval or node.value > maxval:
            return False
        return self._isBST(node.left, minval, node.value) and self._isBST(node.right, node.value, maxval)
    def mirror(self, T_root):
        return self._mirror(T_root)
    def _mirror(self, node):
        if node:
            node.left = self._mirror(node.left)
            node.right = self._mirror(node.right)
            
            temp = node.left
            node.left = node.right
            node.right = temp
        return node
    def sameTree(self, root1, root2):
        return self._sameTree(root1, root2)
    def _sameTree(self, node1, node2):
        if node1 is None and node2 is None:
            return True
        if node1 is not None and node2 is not None:
            return node1.value == node2.value and self._sameTree(node1.left, node2.left) and self._sameTree(node1.right, node2.right)
        return False
    def isFoldable(self, root):
        if root is None:
            return True
        return self._isFoldable(root.left, root.right)
    def _isFoldable(self, left, right):
        if left is None and right is None:
            return True
        if left is None or right is None:
            return False
        return self._isFoldable(left.left, right.right) and self._isFoldable(left.right, right.left)
    def pre_order_iter(self, root):
        ret = []
        stack = []
        node = root
        stack = [node]
        while stack:
            node = stack.pop()
            if node:
                ret.append(node.value)
                stack.append(node.right)
                stack.append(node.left)
        return ret
    def in_order_iter(self, root):
        ret = []
        stack = []
        node = root
        while True:
            while (node is not None):
                stack.append(node)
                node = node.left
            if len(stack) == 0:
                break
#                 return ret
            node = stack.pop()
            ret.append(node.value)
            node = node.right
        return ret
    def post_order_iter(self, root):
        ret = []
        stack = []
        node = root
        stack = [(node, False)]
        while stack:
            node, visited = stack.pop()
            if node:
                if visited:
                    # add to result if visited
                    ret.append(node.value)
                else:
                    # post-order
                    stack.append((node, True))
                    stack.append((node.right, False))
                    stack.append((node.left, False))
        return ret
    def level_order_iter(self, root):
        ret = []
        level = [root]
        while level:
            nextlevel = []
            for node in level:
                if node.left:
                    nextlevel.append(node.left)
                if node.right:
                    nextlevel.append(node.right)
                ret.append(node.value)
            level = nextlevel
        return ret
    def add_iter(self, root, key):
        newNode = Node(key)
        if root is None:
            return newNode
        current = root
        parent = None
        while current:
            parent = current
            if (key == current.value):
                return root
            if (key < current.value):
                current = current.left
                if (current is None):
                    parent.left = newNode
                    return root
            else:
                current = current.right
                if (current is None):
                    parent.right = newNode
                    return root
    def pre_in_build(self, preorder, inorder):
        if not inorder or not preorder:
            return None
        ind = inorder.index(preorder.pop(0))
        root = Node(inorder[ind])
        root.left = self.pre_in_build(preorder, inorder[:ind])
        root.right = self.pre_in_build(preorder, inorder[ind+1:])
        return root
    def post_in_build(self, postorder, inorder):
        if not inorder or not postorder:
            return None
        ind = inorder.index(postorder.pop())
        root = Node(inorder[ind])
        root.right = self.post_in_build(postorder, inorder[ind+1:])
        root.left = self.post_in_build(postorder, inorder[:ind])
        return root
    def sortedarrayBST(self, nums):
        if not nums:
            return None
        mid = len(nums) // 2
        root = Node(nums[mid])
        root.left = self.sortedarrayBST(nums[:mid])
        root.right = self.sortedarrayBST(nums[mid+1:])
        return root
    def path_sum_target(self, root, target):
        if not root:
            return None
        res = []
        lst = []
        self.pathsum_helper(lst, res, root, target)
        return res
    def pathsum_helper(self, temp, output, node, remain):
        if not node.left and not node.right and remain == node.value:
            temp.append(node.value)
            output.append(temp)
        if node.left:
            self.pathsum_helper(temp+[node.value], output, node.left, remain-node.value)
        if node.right:
            self.pathsum_helper(temp+[node.value], output, node.right, remain-node.value)
    
#     def path_sum_target_plus(self, root):
#         if not root:
#             return None
#         res = []
#         lst = []
#         self.pathsum_helper_plus(lst, res, root)
#         return res
#     def pathsum_helper_plus(self, temp, output, node):
#         output.append(temp[:])
#         node1, node2 = node, node
#         while node:
#             value = node.value
#             if node.left:
#                 temp.append(value)
#                 node = node.left
#                 self.pathsum_helper_plus(temp, output, node1)
#                 temp.pop()
#             if node.right:
#                 temp.append(value)
#                 node = node.right
#                 self.pathsum_helper_plus(temp, output, node2)
#                 temp.pop()
```
