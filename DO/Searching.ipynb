```python
# 排列组合
def perm(nums):
    lst = []
    result = []
    perm_helper(lst, result, nums)
    return result
def perm_helper(lst, result, nums):
    if len(nums) == 0:
        result.append(lst[:])
    for i, num in enumerate(nums):
        perm_helper(lst + [num], result, nums[:i] + nums[i+1:])
# perm([1,2,3])
# 有重复元素的情况
def perm_dup(nums):
    lst = []
    result = []
    nums.sort()
    perm_dup_helper(lst, result, nums)
    return result
def perm_dup_helper(lst, result, nums):
    if len(nums) == 0:
        result.append(lst[:])
    for i, num in enumerate(nums):
        if i != 0 and nums[i] == nums[i-1]:
            continue
        perm_dup_helper(lst + [num], result, nums[:i] + nums[i+1:])
# perm_dup([1,2,3,3])
# 只取前k个数
def perm_K(nums, k):
    lst = []
    result = []
    perm_K_helper(lst, result, nums, k)
    return result
def perm_K_helper(lst, result, nums, pos):
    if pos == 0:  # 这个地方就不是nums的长度等于零
        result.append(lst[:])
    for i, num in enumerate(nums):
        perm_K_helper(lst + [num], result, nums[:i] + nums[i+1:], pos - 1)
# perm_K([1,2,3,4],2)
# 找子集
def subset(nums):
    lst = []
    result = []
    subset_helper(lst, result, nums, 0)
    return result
def subset_helper(lst, result, nums, pos):
    result.append(lst[:])
    for i in range(pos, len(nums)):
        lst.append(nums[i])
        subset_helper(lst, result, nums, i + 1)
        lst.pop()
# subset([1,2,3])
# 找子集有重复数字的场景
def subset_dup(nums):
    lst = []
    result = []
    nums.sort()
    subset_dup_helper(lst, result, nums, 0)
    return result
def subset_dup_helper(lst, result, nums, pos):
    result.append(lst[:])
    for i in range(pos, len(nums)):
        if i != pos and nums[i] == nums[i-1]:
            continue;
        lst.append(nums[i])
        subset_dup_helper(lst, result, nums, i + 1)
        lst.pop()
# subset_dup([1,2,3,3])
```

```python
# 子集变种-不定sum问题，找出nums中加起来等于target的子集
# 当然蛮力法可以先把子集都找出来，然后挨个检查
def subset_sum(nums, target):
    lst = []
    result = []
    nums.sort()
    subset_sum_helper(lst, result, nums, 0, target)
    return result
def subset_sum_helper(lst, result, nums, pos, remain):
    if remain == 0:
        result.append(lst[:])
    if remain < 0:
        return
    for i in range(pos, len(nums)):
        if i > pos and nums[i] == nums[i-1]:
            continue
        lst.append(nums[i])
        subset_sum_helper(lst, result, nums, i+1, remain - nums[i])
        lst.pop()
# subset_sum([1,2,3,4], 5)
# 八皇后问题
def check_pos(nums, n):
    for i in range(n):
        if abs(nums[n] - nums[i]) == n - i or nums[n] == nums[i]:  # 绝对值
            return False
    return True
def eight_Q(n):
    lst = []
    result = []
    nums = [-1] * n
    eight_Q_helper(lst, result, nums, 0)
    return result
def eight_Q_helper(lst, result, nums, pos):
    if len(lst) == len(nums):
        result.append(lst[:])
        return   # 因为八皇后不会pop，所以必须找完了就return
    for i in range(len(nums)):
        nums[pos] = i
        if check_pos(nums, pos):
            eight_Q_helper(lst + [i], result, nums, pos+1)
```

```python
# 括号卡特兰数
def gene(n):
    lst = []
    result = []
    gene_helper(lst, result, n, n)
    return result
def gene_helper(lst, result, left, right):
    if right == 0:
        result.append(lst[:])
    if left < right:
        gene_helper(lst + [')'], result, left, right - 1)
    if left > 0:
        gene_helper(lst + ['('], result, left - 1, right)
    return result
# gene(4)
```

```python
# 二分搜索
def binary_search(nums, target):
    if not nums:
        return -1
    left, right = 0, len(nums) - 1
    while left + 1 < right:
        mid = left + (right - left) // 2
        if nums[mid] < target:
            left = mid
        elif nums[mid] > target:
            right = mid
        elif nums[mid] == target:
            return mid
    if nums[left] == target:
        return left
    if nums[right] == target:
        return right
    return -1
# binary_search([1,2,3,4,5,6,7],1)
def search_rota_min(alist):
    if len(alist) == 0:
        return -1    
    left, right = 0, len(alist) - 1
    while left + 1 < right: 
        if (alist[left] < alist[right]):
            return alist[left];
        mid = left + (right - left) // 2
        if (alist[mid] > alist[left]):
            left = mid 
        else:
            right = mid
    if alist[left] < alist[right]:
        return alist[left]
    if alist[left] > alist[right]:
        return alist[right]
    return -1
# search([1,2,3,4,5,6,7,8,0])
def serach_rota_target(nums, target):
    if not nums:
        return -1
    left, right = 0, len(nums) - 1
    while left + 1 < right:
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] > nums[left]:
            if target <= nums[mid] and target >= nums[left]:
                right = mid
            else:
                left = mid
        else:
            if target <= nums[right] and target >= nums[mid]:
                left = mid
            else:
                right = mid
    if nums[left] == target:
        return left
    if nums[right] == target:
        return right
    return -1
# serach_rota_target([5,6,7,8,9,1,2,3,4], 9)  
def search_rank(nums, insert):
    if len(nums) == 0:
        return 0
    left, right = 0, len(nums) - 1
    while left + 1 < right:
        mid = left + (right - left) // 2
        if nums[mid] == insert:
            return mid
        elif nums[mid] < insert:
            left = mid
        else:
            right = mid
    if nums[left] >= insert:
#         print(left, right)
        return left
    if nums[right] >= insert:
        return right
    return right + 1
search_rank([1,2,3,4,5,6,7],5.9)
# 这个规则一番搜索后只left和right是相邻的两个id，也就是说如果left对应的nums[left]大于等于插入值
```
